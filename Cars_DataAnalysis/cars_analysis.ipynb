### 1. Instruction ( For Data Cleaning ) - Find all Null Values in the dataset. If there is any null value in any column, then fill it with the mean of that column.

```python
df_car.isnull()
```

```python
# We are going to use the isnum() and sum() to sum all true cases
df_car.isnull().sum()
```

```python
# But we can only get the mean for datas that should be int or flot type: MSRP, Invoice, Engine Size, 
# Cylinders, Housepower, MPP_City, MPG_Highway, Weight, WeelBase and Length
df_car.head()
```

```python
# Firt let's do it for the collumns that are with the correct data type
df_car['EngineSize'].fillna(df_car['EngineSize'].mean(), inplace= True)
```

```python
# Now Eginesize = 0, there isn't null values
df_car.isnull().sum()
```

```python
# Let's do it for all int and float with null values
df_car['Cylinders'].fillna(df_car['Cylinders'].mean(), inplace= True)
df_car['Horsepower'].fillna(df_car['Horsepower'].mean(), inplace= True)
df_car['MPG_City'].fillna(df_car['MPG_City'].mean(), inplace= True)
df_car['MPG_Highway'].fillna(df_car['MPG_Highway'].mean(), inplace= True)
df_car['Weight'].fillna(df_car['Weight'].mean(), inplace= True)
df_car['Wheelbase'].fillna(df_car['Wheelbase'].mean(), inplace= True)
df_car['Length'].fillna(df_car['Length'].mean(), inplace= True)
```

```python
df_car.isnull().sum()
```

```python
# Converting MSRP and Invoice to float
df_car['MSRP'].describe()
```

```python
#MSRP
msrp = df_car['MSRP'].replace('[\$,)]','', regex=True).replace('[,]','', regex=True).astype(float)
df_car['MSRP'] = msrp
df_car['MSRP'].describe()
```

```python
#Invoice
invoice = df_car['Invoice'].replace('[\$,)]','', regex=True).replace('[,]','', regex=True).astype(float)
df_car['Invoice'] = invoice
df_car['Invoice'].describe()
```

```python
# Filling up null values with the means of each collumns: MSRP and Invoice
df_car['MSRP'].fillna(df_car['MSRP'].mean(), inplace=True)
df_car['Invoice'].fillna(df_car['Invoice'].mean(), inplace=True)
```
